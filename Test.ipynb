```r
library(ggplot2)
library(plyr)
library(dplyr)
```

```r
data <- read.csv("https://github.com/MeSko/test/raw/master/data.csv", 
    sep = ",", na.strings = "NA", strip.white = TRUE, stringsAsFactors = FALSE)
```

# Mean

```r
mean(data$data1, na.rm=TRUE)
mean(data$data2, na.rm=TRUE)
mean(data$data3, na.rm=TRUE)
mean(data$data4, na.rm=TRUE)
```

# Standard deviation

```r
sd(data$data1, na.rm=TRUE)
sd(data$data2, na.rm=TRUE)
sd(data$data3, na.rm=TRUE)
sd(data$data4, na.rm=TRUE)
```

```r
ggplot(data, aes(x=data1))+
  geom_histogram(color="darkblue", fill="lightblue", binwidth=1)
```

```r
ggplot(data, aes(x=data2))+
  geom_histogram(color="darkblue", fill="lightblue", binwidth=1)
```

```r
ggplot(data, aes(x=data3))+
  geom_histogram(color="darkblue", fill="lightblue", binwidth=1)
```

```r
ggplot(data, aes(x=data4))+
  geom_histogram(color="darkblue", fill="lightblue", binwidth=1)
```
